```python
import gradio as gr
import tensorflow as tf
import numpy as np
import requests
```

```python
from tensorflow.keras.applications import InceptionV3 # Training with the imagenet dataset

inception = InceptionV3()
```

```python
# Human readable labels we should save it in the txt file. HEre we have 1000 classes names

response = requests.get("https://git.io/JJkYN")
labels = response.text.split('\n')
```

```python
labels
```

```python
def image_classification(img):
  img = img.reshape((-1, 299, 299, 3))
  img = tf.keras.applications.inception_v3.preprocess_input(img)
  pred = inception.predict(img).flatten()
  return {labels[i]: float(pred[i]) for i in range(1000)}

image = gr.inputs.Image(shape=(299, 299))
label = gr.outputs.Label(num_top_classes=3)

gr.Interface(fn=image_classification, inputs=image, outputs=label, capture_session=True).launch()
```
